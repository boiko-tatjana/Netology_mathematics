#### Центральная предельная теорема

В этом задании нужно убедиться, что ЦПТ действительно работает

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
x = np.linspace(0,10,1001)
rv = sts.expon(2)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,10,1001)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
number_list = [10, 20, 30, 50, 100, 200, 300, 500, 1000]

samples_count = 1000

for index_1 in number_list:
    a = np.array([])
    for index_2 in range(samples_count):
        sample = rv.rvs(size = index_1)
        a = np.append(a, sample.mean())
        
    label = 'sample of ' + str(index_1) + ' histogram'
    plt.axis([2, 4, 0, 12])
    plt.hist(a, density = True, label = label, bins = 30)

    plt.legend()
    plt.ylabel('samples means')
    plt.xlabel('$x$')

    plt.grid()
    plt.show()
```
